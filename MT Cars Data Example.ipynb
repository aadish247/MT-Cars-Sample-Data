```python
import pandas as pd
```

```python
automobile = pd.read_csv(r'C:/Users/aadis/Downloads/mtcars-1.csv')
```

```python
automobile.head()
```

```python
automobile.isnull().sum()
```

```python
X= automobile[['mpg','cyl','disp','hp','drat','wt','qsec','gear','carb']]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
automobile_std = scaler.fit_transform(X)
automobile_std
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
linked = linkage(automobile_std, 'average')
plt.figure(figsize=(15, 10))
dendrogram(linked,
 orientation='top',
 labels = np.array(automobile['model']),
 distance_sort='descending')
plt.ylabel('Euclidean Distance')
plt.title('Dendrogram using Average Linkage')
plt.show()
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
linked = linkage(automobile_std, 'ward')
plt.figure(figsize=(15, 10))
dendrogram(linked,
 orientation='top',
 labels = np.array(automobile['model']),
 distance_sort='descending')
plt.ylabel('Euclidean Distance')
plt.title('Dendrogram using Ward Linkage')
plt.show()
```
